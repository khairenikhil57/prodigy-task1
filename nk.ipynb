```python
plt.figure(figsize = (8,4))

ax1= sns.countplot(x = 'Survived', hue = 'Sex', data = df, palette ='Blues') 
legend_labels,_= ax1. get_legend_handles_labels() 
ax1. legend (bbox_to_anchor=(2,4)) 
plt.title('survived distribution with sex', size = 20)
plt.xlabel('distribution')
plt.ylabel('Survived')

plt.legend(['male', 'femail'])

plt.show()
```

```python
cancelled_perc = df["Sex"].value_counts(normalize = True)
print (cancelled_perc)
plt.figure(figsize = (5,4))
plt.title('Count')
plt.bar(['male', 'femail'],df['Sex'].value_counts(), edgecolor='k', width = 0.7) 
plt.show()
```
